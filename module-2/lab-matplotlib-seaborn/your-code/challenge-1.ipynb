#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1,2,1)
plt.plot(x,y,lw=3,color="b", ls='--')
plt.subplot(1,2,2)
plt.plot(x,z,lw=3, color="r",ls='-')
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig,axes=plt.subplots(2,1)
axes[0].plot(x,y,color="b", lw=3, ls='--')
axes[1].plot(x,z,color="r", lw=3, ls='-')
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[1].plot(x,z,color="red", lw=3, ls='--')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2,x, np.exp(x))
axes[0].set_xscale("linear")
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2,x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)")
```

#### In the vehicles data set that you have downloaded, conduct exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. Please provide at least one inssight from the plot that you have drawn

```python
vehicles = pd.read_csv('vehicles.csv')
vehicles.head()
```

Scatter Plot

```python
# your code here
x = vehicles['City MPG']
y = vehicles['Fuel Cost/Year']
plt.scatter(x,y)
plt.xlabel('City MPG')
plt.ylabel('Fuel Cost/Year')
```

Box Whisker Plot

```python
# your code here
x = vehicles['CO2 Emission Grams/Mile']
green_diamond = dict(markerfacecolor='g', marker='D')
plt.boxplot(x, flierprops=green_diamond)
plt.title('CO2 Emission Grams/Mile')
```

Histogram

```python
# your code here
x = vehicles['Fuel Cost/Year']
plt.hist(x)
plt.title('Histogram of Fuel Cost/Year')
```

Bar Chart

```python
ax = sns.barplot(x="Fuel Type", y="Fuel Cost/Year", data=vehicles)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
