```python
#11-Récupération des dates à partir de la colonne "Case Number"
```

```python
#11-a: split des éléments de chaque ligne selon "."
```

```python
data_select["Date_propre"] = data_select["Case Number"].map(lambda x: x.split("."))
```

```python
#Définition d'une fonctione pour traiter les cas où il y a des "-"
def correct_dashes(l):
    if len(l)==1:
        return l[0].split("-")
    elif len(l)==2:
        all_parts_splitted = []
        for part in l :
            part_splited = part.split("-")
            all_parts_splitted.extend(part_splited)
        return all_parts_splitted
    else:
        return l
    
#Définition d'une fonction qui supprime le dernier élément de chaque ligne qui est une liste
def supp_4ème_element(l):
    for x in l:
        if len(l)==4:
            l.pop()
    return l
        
#Ajout des tirets pour reformer une date au bon format
def join_tiret(l):
    for x in l:
        if len(l)==2:
            return l[0]+'-'+l[1]
        if len(l)==3:
            return l[0]+'-'+l[1]+'-'+l[2]
    
    
data_select["Date_propre"] = data_select["Case Number"].map(lambda x: x.split(".")) #split des éléments de chaque ligne selon "."
data_select["Date_propre"] = data_select["Date_propre"].map(correct_dashes) #split des éléments de chaque ligne selon "-"
data_select["Date_propre"] = data_select["Date_propre"].map(supp_4ème_element) #liste pour laquelle le 4ème élément (lettre) est supprimé
data_select["Date_propre"] = data_select["Date_propre"].map(join_tiret) #string contenant le bon format de date
```
