```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
from bs4 import BeautifulSoup
```

```python
url = 'https://api.tronalddump.io/search/quote?query=women'
```

```python
response = requests.get(url)
response
```

```python
results = response.json()
#results
```

```python
flattened_data = json_normalize(results)
flattened_data.head()
#type(flattened_data)
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data


quotes = pd.DataFrame(flattened_data['_embedded.quotes'])
quotes.head()

nested_columns = ['_embedded.quotes']
quotes_fl=flatten(quotes, nested_columns)
quotes_fl
```

```python
e0 = pd.DataFrame(quotes_fl['_embedded.quotes_0'])
e0.head()

nested_columns = ['_embedded.quotes_0']
e0_fl=flatten(e0, nested_columns)
e0_fl = e0_fl[['_embedded.quotes_0_appeared_at', '_embedded.quotes_0_created_at', '_embedded.quotes_0_quote_id', '_embedded.quotes_0_tags', '_embedded.quotes_0_updated_at', '_embedded.quotes_0_value']]
#list(e0_fl)
e0_fl
```
