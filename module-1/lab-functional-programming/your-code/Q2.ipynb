In HTML, there are a lot of messy codes such as HTML tags, Javascripts, [unicodes](https://www.w3schools.com/charsets/ref_utf_misc_symbols.asp) that will mess up your bag of words. We need to clean up those junk before generating BoW.

For instance, you can have a `strip_html_tags` function to remove all HTML tags, a `remove_punctuation` function to remove all punctuation, a `to_lower_case` function to convert string to lowercase, and a `remove_unicode` function to remove all unicodes.

```python
# Define your string handling functions below
# Minimal 3 functions
import re
from string import punctuation

def strip_html_tags(string):         
    string_strip_html = re.sub('<.*?>', '', string)
    
    return string_strip_html
    
def remove_punctuation(string):
    
    return ''.join(c for c in string if c not in punctuation)

    #punctuaction_f = ''.join(re.findall(r'[^\w\s]', string))
    
    #return punctuaction_f
            
    
def to_lower_case(string):
    lower_case = (string.lower())
        
    return lower_case
```

Next, paste your previously written `get_bow_from_docs` function below. Call your functions above at the appropriate place.

```python
# Import required libraries
import re
# Define function
def get_bow_from_docs(docs):
    #string = ''
    corpus = []
    bag_of_words = []
    term_freq = []
     
    for doc in docs:
        with open(doc,"r") as f:
            string = f.read()
        
        string = strip_html_tags(string)
        string = remove_punctuation(string)
        string = to_lower_case(string)
        corpus.append(string)
    
    
    for i in corpus:
        split = (re.split(' ', i))
        for j in split:
            if j not in bag_of_words:
                bag_of_words.append(j)
                    
    for i in corpus:
        split = (re.split(' ', i))
        liste = []
        for j in bag_of_words:
            liste.append(split.count(j))
        term_freq.append(liste)

    
    # Now return your output as an object
    return {
        "bag_of_words": bag_of_words,
        "term_freq": term_freq
    }

docs = ["../../lab-string-operations/your-code/doc1.txt", "../../lab-string-operations/your-code/doc2.txt","../../lab-string-operations/your-code/doc3.txt"]

print(get_bow_from_docs(docs))
```
